```python
import pandas as pd
import numpy as np
import os
import cv2
```

```python
df = pd.read_csv('./data/label.csv')
df
```

```python
with_helmet = list(df[df['label'] == 'With Helmet']['file_name'])
without_helmet = list(df[df['label'] == 'Without Helmet']['file_name'])
```

```python
with_helmet
```

```python
image_list = os.listdir('./data/cut_image')
```

```python
for image in image_list :
    
    img = cv2.imread('./data/cut_image/'+image)
    
    if image in with_helmet :    
        cv2.imwrite('./data/with_helmet/'+image, img)
    
    else :
        cv2.imwrite('./data/without_helmet/'+image, img)
```
