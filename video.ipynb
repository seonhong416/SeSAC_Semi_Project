```python
import os
import cv2

def video2frame(video_name, invideofilename, save_path):
    vidcap = cv2.VideoCapture(invideofilename)
    count = 0
    while True:
      success,image = vidcap.read()
      if not success:
          break
      if(int(vidcap.get(1)) % 10 == 0):
        print ('Read a new frame: ', success)
        fname = "{}.png".format(f"{count}")
        cv2.imwrite(save_path + video_name + '_' + fname, image) # save frame as JPEG file
        count += 1
    print("{} images are extracted in {}.". format(count, save_path))

    
    
```

```python
video_list = os.listdir('./add_data/')

for video_name in video_list :
    name = video_name.split('.')[0]
    video2frame(name, './add_data/' + video_name, './Detection/add_data/')
```

```python
video_list = os.listdir('../Semi_Data/withouthelmet')


for video_name in video_list :
    name = video_name.split('.')[0]
    video2frame(name, '../Semi_Data/withouthelmet/' + video_name, '../Semi_Data/withouthelmetimages/')
```

```python
# 이미지 사이즈 조절하여 저장 하는 함수

def img_resize_save(img_path, save_path, save_img_name) :
    img_list = os.listdir(img_path)
    count = 0
    for img in img_list :
        image = cv2.imread(img_path + img)
        
        image = cv2.resize(image, (416, 416))
        
        cv2.imwrite(save_path+save_img_name+'_'+str(count)+'.png', image)
        count += 1
```

```python
img_resize_save('./data/with_helmet_add/', './data/with_helmet_add_416/', 'with_helmet_add')
```
